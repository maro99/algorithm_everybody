### 1. 용어 정리

친구(일촌): 어떤 두 사람이 직접 아는 사이일때.  
모든친구(친척): 어떤 사람이 직접 아는 친구들과 그 친구들의 친구들. 즉 직간접으로 아는 모든 사람을 말한다.(자신포함)

   ㅣ ㅣ /

1~6 까지의 이름이 붙여진 꼭짓점(vertrex)이 여섯 개 있고 그 꼭지점 사이를 연결하는 선(edge)일곱개가 있다.

### 5. 모든 친구 찾기 알고리즘

tom 의 모든 친구의 출력은 간단하다.
먼저 tom 자신 출력
tom친구 찾아보니 jerry가 있어서 jerry출력

summer의 모든 친구를 출력하는 경우
먼저 summer 자신을 출력
summer의 친구들을 찾아보니 세명이 있다. 세명 출력
이 세명의 친구들을 따라가 본다. 

이처럼 친구가 여러명이고 그 친구들의 친구가 여러명일때는 기억력 만으로는 모든 친구를 따라가기 힘들다.---> 메모를 하자. 


두가지 메모가 필요하다. 

1. 첫째.  
앞으로 처리해야 할 사람들.
꼬리를 물고 나오는 친구의 친구들을 한명도 빠뜨리지 않고 처리하려면 
친구의 이름이 나올 때마다 메모지에 적어두었다가 한 명씩 처리하면서 메모지에서 지워야한다.  


2. 둘재.   
이미 추가된 사람들.
친구 추적 과정에서 한 명이 여러번 나오거나 추적이 무한하게 반복되지 않게 하려면 
이미 처리 대상으로 올린 사람은 중복되지 않도록 메모지에 적어 두어야 한다.

( 앞서서 예에서 tom-jerry에서 두번째 과정을 하지 않으면 
 tom은 jerrt를 친구로 출력하고 jerrt는 다시 tom을 친구로 출력하고 .... 영원히 반복된다.)

 우리까 만뜰 알고리즘에서는 '앞으로 처리해야 할 사람으 넣어 두었다가 다시 하나씩 꺼내기위한 기억장소' 로 
 큐를 이용한다. 또한 '이미 처리 대상으로 추가한 사람들을 적어둘 기억장소로 집합 을 이용해보겠다. 
 
 
 1. 앞으로 처리할 사람을 저장할 큐를 만든다. 
 2. 이미 큐에 추가한 사람을 저장할 집합을 만든다. 
 3. 검색의 출발점이 될 사람을 큐와 집합에 추가한다. 
 4. 큐에서 사람이 남아 있다면 큐에서 처리할 사람을 꺼낸다. 
 5. 꺼낸 사람을 출력한다. 
 6. 꺼낸 사람의 친구들 중 아직 큐 에 추가된 적이 없는 사람을 골라 큐와 집합에 추가한다.
 7. 큐에 처리할 사람이 남아 있다면 4번 과정부터 다시 반복한다. 

  

```python
import copy

#어떤 사람의 모든 친구를 구하는 함수를 만들자. 
# 입력 
# name = 사람이름
# group = 사람들의 전체 집합.

def find_all_friends(name,group):
    # 처리해야될 리스트
    waited_list = []
    
    # 처리한 사람 리스트
    printed_list = []
    
    # loop들어가기전 wait_list에 받은 이름 추가
    waited_list.append(name)
    
    break_count = 0 
    while(1):
        
        break_count += 1
        print(f'itter: {break_count}')
        if break_count > 100:
            break
            
        # 만약 더이상 waited_list에 name이 없다면 
        if not waited_list:
            break
            
            
        
        # wait_list에 있는 name들의 친구들을 모두 일단 waited_list에 추가한다. 
        # 단 printed_list에 그 이름이 없어야 한다. 
        
        waited_list_copy = copy.deepcopy(waited_list)
        print(waited_list_copy)
        for one_name in waited_list_copy:
            
            # 친구 탐색의 주최가 되는 name이 아직 print안됬다면 print해준다.
            if one_name not in printed_list:
                print(one_name)
                printed_list.append(one_name)
            
            # 친구 탐색의 주최가 되는 name은 pop 해서 없애준다. 
            waited_list.pop(waited_list.index(one_name))

            # 친구 탐색의 주최의 친구들을 받아온다.
            his_friends = group[one_name]

            for friend in his_friends:
                # 만약 printed_list에 그 친구가 없다면
                if friend not in printed_list:
                    print(friend)
                    printed_list.append(friend)
                    waited_list.append(friend)
                    # 여기서 waited_List에 사람을 추가해 주는데 어짜피 다음 itter 에서 
                    # printed_list에 이사람 이름이 있기 때문에 인쇄는 안되고 이사람의 친구들만 탐색해서 괜찮다.
        print('@@@@@@@@@@@2')
            
        
        
```

```python
# 책에서는 이렇게 했다. 

# 대상을 pop해서 변수에 담고 그것에 친구에 대한 loop돌았다.  

def find_all_friends(name,group):
    # 처리해야될 리스트
    waited_list = []
    
    # 처리한 사람 리스트(여기서 처리란 waited_list에 한번 추가됬던것을 말한다. )
    done_list = []
    
    # loop들어가기전 wait_list에 받은 이름 추가
    waited_list.append(name)
    done_list.append(name)
    
    
    
    # 나같은 경우에서는 전번에서 추가된 name들에 대해서 모두 loop를 돌았지만
    # 책에서는 waited_list의 하나의 이름에 대해서 loop를 돌고 
    #그 친구들을 waited_List에 추가까지만 한 loop돈다. 
    while(waited_list):
        
        one_name = waited_list.pop(0)
        print(one_name)
        
        for friend in group[one_name]:
            if friend not in done_list:
                waited_list.append(friend)
                done_list.append(friend)
        
```

### 6. 친밀도 계산 알고리즘

A와 B가 친구이고 B와 C가 친구라고 가정해본다. (A-B-C)  
A를기준으로 B의 친밀도는 1  
A를 기준으로 C의 친밀도는 2  

일반적으로 A라는 사람과 X라는 사람의 친밀도가 n이면 X의 친구 Y는 A와 친밀도가 n+1 이 된다.   
이 성질을 이용하여 어떤 사람의 친구들을 큐에 넣을 때 친밀도를 1씩 증가시키면 된다. 



```python
# 책에서는 이렇게 했다. 

# 대상을 pop해서 변수에 담고 그것에 친구에 대한 loop돌았다.  

def find_all_friends(name,group):
    # 처리해야될 리스트
    waited_list = []
    
    # 처리한 사람 리스트(여기서 처리란 waited_list에 한번 추가됬던것을 말한다. )
    done_list = []
    
    # 친밀도 
    how_close = 0 
    
    # loop들어가기전 wait_list에 받은 이름 추가
    waited_list.append((name,how_close))
    done_list.append(name)

    
    
    
    # 나같은 경우에서는 전번에서 추가된 name들에 대해서 모두 loop를 돌았지만
    # 책에서는 waited_list의 하나의 이름에 대해서 loop를 돌고 그 친구들을 waited_List에 추가까지만 한 loop돈다. 
    while(waited_list):
        
        one_name,close_rate = waited_list.pop(0)
        print(one_name,close_rate)
        
        for friend in group[one_name]:
            if friend not in done_list:
                waited_list.append((friend,close_rate+1))
                done_list.append(friend)
        
```
